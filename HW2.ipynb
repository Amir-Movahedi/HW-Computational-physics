```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constant 
g = 9.8
l = 20 #cm
theta0 = 0.03
omega0 = 0

params = [g, l]
state = np.array([theta0, omega0])
# state = [theta, omega])

def Euler_Cromer(state, paramsdt, dt):
    Omega = state[1] - (params[0]/params[1])*state[0]*dt
    Theta = state[0] + Omega*dt
    return np.array([Theta, Omega])





nStep = int(10./0.04)
t = np.linspace(0,100,nStep)
dt = t[1]-t[0]

statelist = []
statelist.append(state)

for i in range(len(t)-1):
    statelist.append(Euler_Cromer(statelist[i], params,dt))
allstates = np.array(statelist)
Theta = allstates[:,0]
Omega = allstates[:,1]
```

```python
plt.plot(t, Omega,'r-')
plt.xlabel(r'$t$')
plt.ylabel(r'$\omega$')
plt.show()

plt.plot(t, Theta,'r-')
plt.xlabel(r'$t$')
plt.ylabel(r'$\theta$')
plt.show()
```
